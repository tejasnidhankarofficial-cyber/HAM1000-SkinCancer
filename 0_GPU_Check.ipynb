```python
import torch

# Define the device variable
# This is the line that checks for Apple Silicon GPU support
if torch.backends.mps.is_available():
    device = torch.device("mps")
    print("✅ PyTorch successfully configured for Apple M-series (MPS) GPU acceleration.")
    print(f"Using device: {device}")
else:
    # Fallback if the support is not detected
    device = torch.device("cpu")
    print("❌ MPS not available. Training will default to CPU.")
    
# Test moving a tensor (data) to the detected device
# All your models and data must be explicitly sent to this device for acceleration.
x = torch.rand(5, 5)
x_on_gpu = x.to(device)

print(f"\nTensor (data) location: {x_on_gpu.device}")
print(f"PyTorch version: {torch.__version__}")
```
